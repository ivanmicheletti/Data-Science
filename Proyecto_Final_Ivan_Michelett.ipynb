##Regex

```python
import re

def resumir_titulo(titulo):
    # Definir patrón regex para extraer la parte relevante del título
    patron = re.compile(r'\b(Data Engineer|Data Science|Data Scientist|Machine Learning|Data Analyst|Data Architect)\b', flags=re.IGNORECASE)

    # Buscar el patrón en el título
    coincidencias = patron.findall(titulo)

    try:
      output=coincidencias[0]
    except:
      output=titulo
    return output
```

```python
salario['titulo_trabajo']=salario.titulo_trabajo.apply(resumir_titulo)
```

```python
salario.titulo_trabajo.value_counts()
```

```python
salario['titulo_trabajo'] = salario['titulo_trabajo'].replace({'Data Analytics': 'Data Analyst',
                                                               'Data Operations Analyst':'Data Analyst',
                                                               'Data Analytics Consultant':'Data Analyst',
                                                               'Data Analytics Lead':'Data Analyst',
                                                               'ML Engineer':'Machine Learning',
                                                               'Data Quality Analyst':'Data Analyst',
                                                               'Data Analytics Manager':'Data Analyst',
                                                               'Data Infrastructure Engineer':'Data Engineer'})
```

```python
salario.titulo_trabajo.value_counts()
```

El data set original tenia categorías de títulos de trabajos muy variadas. Se resumieron los títulos que expresaban el mismo trabajo pero textualmente se escribian diferentes.

##Eliminación de Outliers

```python
#Identificar Outliers
Q1=salario['salario_usd'].quantile(0.25)
Q3=salario['salario_usd'].quantile(0.75)
IQR=Q3-Q1
```

```python
outliers = salario[(salario['salario_usd'] < Q1 - 1.5* IQR) | (salario['salario_usd'] > Q3 + 1.5 * IQR)]
```

```python
#Eliminar Outliers
salario = salario.drop(outliers.index)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor


from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.impute import SimpleImputer


from sklearn.model_selection import cross_val_score

from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=salario
df.head()
```

# Preprocesamiento

```python
X=df.drop(['salario_usd'],axis=1)
```

```python
y=df.salario_usd
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
df.dtypes
```

```python
lista_numeric_features = df.select_dtypes(include=['int64']).drop(['salario_usd'], axis=1).columns
```

```python
lista_numeric_features
```

```python
lista_categorical_features = df.select_dtypes(include=['object']).columns
```

```python
lista_categorical_features
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
```

```python
categorical_transformer = Pipeline(steps=[
      ('imputer', SimpleImputer(strategy='most_frequent')),
      ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
Preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, lista_numeric_features),
        ('cat', categorical_transformer, lista_categorical_features)])
```

```python
Preprocessor
```

```python
Preprocessor.fit_transform(X_train)
```

#Modelado

Me armo un pipe:

```python
pipe = Pipeline(steps=[('preprocessor', Preprocessor),
                      ('modelo', LinearRegression())])
```

```python
pipe
```

Defino la grilla con las diferentes modelos e hiperparámetros:

```python
params_grid = [

    {
       "modelo":[LinearRegression()]
    },


    {
     "modelo" :[DecisionTreeRegressor()],
     "modelo__max_depth": [10, 20, 30]
    },


    {
     "modelo" :[RandomForestRegressor()],
     "modelo__n_estimators": [50, 100, 200]
    },


    {
     "modelo" :[XGBRegressor()],
     "modelo__n_estimators": [50, 100, 200],
     "modelo__max_depth": [3, 6, 9],
     "modelo__learning_rate": [0.01, 0.1, 0.3],

    },

    {
     "modelo" :[CatBoostRegressor()],
     "modelo__n_estimators": [500, 1000, 2000],
     "modelo__learning_rate": [0.01, 0.03, 0.06]
    },

    {
     "modelo" :[LGBMRegressor()],
     "modelo__n_estimators": [50, 100, 200],
     "modelo__learning_rate": [0.01, 0.1, 1]
    },


]
```

Instancio utilizando la grilla y cross validation:

```python
pipe_search_cv=GridSearchCV(pipe, params_grid, cv=5, verbose=0, n_jobs=4)
```

Realizo el entrenamiento y observo el mejor modelo con los hiperparametros:

```python
pipe_search_cv.fit(X_train, y_train)
#el pipeline ganador
pipe_search_cv.best_estimator_
```

```python
pipe_search_cv.best_params_
```

```python
pipe_final=pipe_search_cv.best_estimator_
```

```python
pipe_final.fit(X_train,y_train)
```

```python
y_pred_test=pipe_search_cv.predict(X_test)
```

```python
#Calculo la métrica para validar el modelo:

r2 = r2_score(y_test, y_pred_test)
print(f"R2 Score: {r2}")
```
